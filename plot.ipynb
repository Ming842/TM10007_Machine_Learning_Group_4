```python
import matplotlib.pyplot as plt

# Plot sensitivity against n_estimators
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# Flatten the axes array for easier iteration
axes = axes.flatten()

columns = ['n_estimators', 'min_samples_split', 'min_samples_leaf', 'max_samples', 'max_depth', 'bootstrap']
for i, col in enumerate(columns):
    axes[i].scatter(csv[col], csv['auc'], alpha=0.5)
    axes[i].set_title(f'AUC vs {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('AUC')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```
