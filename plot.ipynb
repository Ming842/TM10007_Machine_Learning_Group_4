```python
import pandas as pd
import numpy as np  

csv = pd.read_csv(r'E:\OneDrive\School\Technical Medicine\TM Jaar 1\Machine learning\TM10007_Machine_Learning_Group_4\TM\best_hyperparameters.csv')
```

```python
import matplotlib.pyplot as plt

# Plot sensitivity against n_estimators
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# Flatten the axes array for easier iteration
axes = axes.flatten()

columns = ['n_estimators', 'min_samples_split', 'min_samples_leaf', 'max_samples', 'max_depth', 'bootstrap']
for i, col in enumerate(columns):
    means = csv.groupby(col)['sensitivity'].mean()
    stds = csv.groupby(col)['sensitivity'].std()
    errors = stds / np.sqrt(csv.groupby(col)['sensitivity'].count())
    
    axes[i].bar(means.index, means, yerr=errors, alpha=0.5, capsize=5)
    axes[i].set_title(f'Sensitivity vs {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Sensitivity')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```
